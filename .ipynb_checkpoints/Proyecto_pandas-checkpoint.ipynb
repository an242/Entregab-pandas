```python
#Ejercicio 1 

import pandas as pd

```

```python
df_sample = pd.read_csv('/Users/joelalejandro/Desktop/ProyectoPandas/sample_NYC_parking_issues.csv', sep=";")
df_sample = df_sample.dropna(axis=1, how='all')
# Rellenar los NaN con un valor específico, por ejemplo, 0
df_sample = df_sample.fillna(0)

print(df_sample)

```

```python
#Ejercicio 2 

print(df_sample.head())

# Primero, asegurémonos de que la columna 'Vehicle Year' esté en formato numérico.
# Convertir la columna a números, forzando los errores a NaN
df_sample['Vehicle Year'] = pd.to_numeric(df_sample['Vehicle Year'], errors='coerce')

# Ahora eliminar las filas donde 'Vehicle Year' tiene valores NaN o fuera del rango de años válidos (1900-2024)
df_cleaned = df_sample[(df_sample['Vehicle Year'] >= 1900) & (df_sample['Vehicle Year'] <= 2024)]

# Verificar el DataFrame limpio
print(df_cleaned.head())

# Convertir 'Vehicle Year' a enteros
df_cleaned['Vehicle Year'] = df_cleaned['Vehicle Year'].astype(int)

# Verificar el resultado
print(df_cleaned.head())







```

```python
print(df_cleaned['Vehicle Year'].isna().sum())
# Verificar que todos los valores de 'Vehicle Year' están dentro del rango válido
print(df_cleaned['Vehicle Year'].min())  # El valor mínimo
print(df_cleaned['Vehicle Year'].max())  # El valor máximo
```

```python
#Ejercicio 3 

#Primero busco dentro de la columna los nosmbres de los vehículos y eligo los 4 tipos por los que los quiero filtrar.
print(df_cleaned['Vehicle Make'])
valid_vehicle_makes = ['DODGE','HONDA','TOYOT','FORD']
# Filtrar el DataFrame para mantener solo las filas donde 'Vehicle Make' esté en la lista de marcas válidas
df_cleaned = df_sample[df_sample['Vehicle Make'].isin(valid_vehicle_makes)]

# Verificar cuántas filas se eliminaron
rows_deleted = len(df_sample) - len(df_cleaned)

# Mostrar el resultado
print(f'Número de filas eliminadas: {rows_deleted}')
print(f'Número de filas restantes: {len(df_cleaned)}')
print(df_cleaned.head())



```

```python
#Ejercicio 4 

import matplotlib.pyplot as plt

# Contar la cantidad de vehículos por marca ('Vehicle Make')
vehicle_counts = df_cleaned['Vehicle Make'].value_counts()

# Crear el gráfico de barras
plt.figure(figsize=(10, 6))  # Establecer el tamaño de la figura
vehicle_counts.plot(kind='bar', color='skyblue')

# Crear el gráfico de barras
plt.figure(figsize=(10, 6))  # Establecer el tamaño de la figura
vehicle_counts.plot(kind='bar', color='skyblue')

# Añadir título y etiquetas
plt.title('Cantidad de Vehículos por Marca', fontsize=16)
plt.xlabel('Marca de Vehículo', fontsize=12)
plt.ylabel('Cantidad de Vehículos', fontsize=12)

```

```python
#Ejercicio 5 


color_mapping = {
    'wh': 'white', 'wh\\': 'white', 'WH': 'white', 'white': 'white',
    'bl': 'blue', 'BL': 'blue', 'blue': 'blue',
    'gr': 'green', 'GR': 'green', 'green': 'green',
    'rd': 'red', 'RD': 'red', 'red': 'red',
    'bk': 'black', 'BK': 'black', 'black': 'black',
    'yl': 'yellow', 'YL': 'yellow', 'yellow': 'yellow',
    'gray': 'grey', 'GRY': 'grey', 'GREY': 'grey', 'grey': 'grey',
    'org': 'orange', 'ORG': 'orange', 'orange': 'orange',
    'br': 'brown', 'BR': 'brown', 'brown': 'brown',
    # Agregar más colores según sea necesario...
}

# Asegurarse de que la columna 'Vehicle Color' no tiene valores nulos
df_sample['Vehicle Color'] = df_sample['Vehicle Color'].fillna('')

# Normalizar la columna 'Vehicle Color' usando el diccionario de mapeo
df_sample['Vehicle Color'] = df_sample['Vehicle Color'].apply(
    lambda x: color_mapping.get(str(x).strip().lower(), str(x).strip().lower())
)

# Verificar los primeros registros para asegurarnos de que los valores se han unificado
print(df_sample['Vehicle Color'].head())

# Verificar los valores únicos después de la normalización
print(df_sample['Vehicle Color'].unique())











```

```python
#Ejercicio 6 

import matplotlib.pyplot as plt

# Contamos cuántos vehículos de cada marca tienen un determinado color
color_by_make = df_cleaned.groupby(['Vehicle Make', 'Vehicle Color']).size().unstack(fill_value=0)

# Verificar la tabla resultante para asegurarnos de que tiene el formato adecuado
print(color_by_make)

# Crear un gráfico de barras apiladas
color_by_make.plot(kind='bar', stacked=True, figsize=(12, 7), colormap='tab20')

# Añadir título y etiquetas
plt.title('Cantidad de Vehículos por Marca y Color', fontsize=16)
plt.xlabel('Marca de Vehículo', fontsize=12)
plt.ylabel('Cantidad de Vehículos', fontsize=12)

# Mostrar leyenda con los colores
plt.legend(title='Color de Vehículo', bbox_to_anchor=(1.05, 1), loc='upper left')

# Mostrar el gráfico
plt.tight_layout()
plt.show()






















```
